# Vektorrechnung 

Dieses Notebook enthällt die Grundlagen der Vektorrechnung wie sie für das Verständnis der klassischen Mechanik notwendig sind. 

Gemeinhin sind Zahlen wenn sie alleine stehen nichts weiter als Skalare, sie besitzen zwar einen Wert, dieser hat jedoch keine Richtung. 

Überlegt man sich was geschieht wenn man einen Ball wirft wird schnell klar, dass um die Flugbahn, die Geschwindigkeit und andere Größen sinnvoll beschreiben zu können, eine solche ungerichtete Größe nicht geeignet ist. 

Hier schaffen Vektoren abhilfe. Sie können in einem beliebig dimensionalen Raum einem Skalar mit einer Richtung versehen. Die Länge des Vektors, oder auch Betrag, ist analog zu einem Skalar. 

## Definition

Es seien $A$ und $B$ beliebige, aber eindeutige Punkte in einem beliebig dimensionalen Raum (bspw. zweidimensional als Ebene oder dreidimensional als Volumen). Zwischen diesen Punkten kann der Vektor $\vec{AB}$ als Pfeil symbolisch dargestellt werden. Hierdurch wird die Richtung des Vektors eindeutig mit Anfangspunkt $A$ und Endpunkt $B$, und dem Betrag $|\vec{AB}|$ = $a$ definiert. 

Konventionsgemäß wird der Anfangspunkt eines Vektors als erstes Symbol und der Endpunkt als zweites Symbol verwendet. 

## Interpretation

Vektoren können verschiedene Bedeutungen besitzen, da sie für sich genommen, nur ein im Raum gerichteten Betrag darstellen. So kann man einen Vektor definieren welcher den Betrag und die Richtung einer Kraft definiert, wie bspw. die Lagerreaktion in den Stützen einer Brücke, aber auch die Geschwindigkeit (Betrag) und die Richtung mit der sich ein Fußball übers Spielfeld bewegt. 

Im Rahmen dieser Vorlesung wird sich auf zwei- und dreidimensionale Problemstellungen konzentriert. 

## Eigenschaften

Vektoren verhalten sich zum Großteil ähnlich wie skalare Größen, durch die Eigenschaft, dass sie sich im Raum befinden kommen jedoch neue Eigenschaften hinzu. Vektoren können nicht nur im Betrag im Bezug zueinenader gesetzt werden, sondern auch in ihrer Ausrichtung (Parallel und Antiparallel) sowie ihre Lage zueinenader im Raum (identisch, geschnitten oder windschief). 

Die Rechenregeln für Vektoren werden im Folgenden vorgestellt. 

### Vektoraddition 

Für die Vektoraddition werden zwei Vektoren $\vec{AB}$ und $\vec{BC}$ angenommen. Also Vektoren die von einem beliebigen Punkt $A$ zu einem Punkt $B$ reichen, welcher als Ende für $\vec{AB}$ und als Anfang für $\vec{BC}$ dient. Addiert man diese Vektoren resultiert der Vektor $\vec{AC}$. 

\begin{equation}
\vec{AB} + \vec{BC} = \vec{AC}
\end{equation}

Diese Formel ist recht einfach zu interpretieren. Der Leser stelle sich vor, das die Punkte $A$, $B$ und $C$ durch Fähnchen auf einer Wiese markierte Stellen sind. Läuft man von Fähnchen $A$ über Fähnchen $B$ hin zu Fähnchen $C$ kommt man gleichermaßen am Ziel an als wenn man direkt von $A$ nach $C$ gelaufen wäre. 

### Kummutativgesetz

Wie auch bei Skalaren ist die Reihenfolge der Additionen für Vektoren beliebig, es gilt

\begin{equation}
    \vec{a} + \vec{b} = \vec{b} +\vec{a}
\end{equation}

Aufgezeichnet bilden die Vektoren ein Parralelogram, also eine geometrische Figur dessen gegenüberliegende Seiten parallel und von gleicher Länge sind. Auch hier ist schnell zu sehen, dass der oben beschriebene Zusammenhang wahr ist. 

Interpretiert man das Parralelogram als Wegstrecke um ein gleichseitiges Gebäude, ist es egal ob man an der ersten Ecke nach links oder rechts geht, Ziel und zurückgelegte Strecke sind identisch. 

### Assoziativgesetz 

Hier wird die Vektoraddition verallgemeinert auf eine beliebige Anzahl von Vektoren die in beliebiger Reihenfolge zu einem resultierenden Vektor verkettet werden. 

\begin{equation}
    \vec{a}+(\vec{b}+\vec{c}) = \vec{AB} + \vec{BC} = (\vec{a}+\vec{b})+\vec{c}
\end{equation}

#### Erste Schritte Richtung Physik: Das Kräfteparallelogram

Um die vorgestellten Gesetze das erste mal in die Praxis anzuwenden, soll nun das Kräfteparallelogramm betrachtet werden. Um einigen Inhalten vorwegzugreifen: Wenn mehrere Kräfte im gleichen Punkt angreifen, können diese zu einer resultierenden Kraft zusammengefasst werden, ohne das Verhalten des Systems zu ändern. 

Die graphische Lösung dieses Problems ist die Zeichnung eines Kräfteparallelogramms. Hier werden alle in einem Punkt angreifenden Kräfte eingezeichnet, und folgend so verschoben, dass sie graphisch addiert werden. Verbindet man den Angriffspunkt der Kräfte mit dem Endpunkt der Vektorkette erhällt man Betrag und Richtung der resultierenden Kraft. 



### Vektorsubtraktion 

Die Vektorsubtraktion ist, wenn man versucht sie sich im ersten Moment vorzustellen, vielleicht etwas unintuitiv, da man versucht sich eine negative Strecke vorzustellen. Wer jemals einen Zollstock in der Hand hielt weis jedoch, man kann nur in positive Längen messen. 

Wie bildet man nun eine negative Strecke ab? 

Die Methode ist simpel, anstatt eine Strecke $\vec{AB}$ in der vorgegebenen Richtung von Punkt $A$ nach Punkt $B$ zu gehen vertauscht man Anfangs- und Endpunkpunkte. Hiermit gilt

\begin{equation}
    -\vec{AB} = \vec{BA}
\end{equation}

wobei der Vektor $\vec{BA}$ als inverser Vektor zu $\vec{AB}$ bezeichnet wird. 

Ab diesem Punkt verhällt sich die Subtraktion der Vektoren wie die Addition, als eine Verkettung der Vektoren miteinander.

### Nullvektor

Der Nullvektor ist ein Sonderfall der Vektoren, dessen Betrag $|\vec{0}|$ = 0 ist und keine Richtung besitzt. (Oder jede Richtung, je nachdem wen man fragt)

Mit den eben vorgestellten Regeln gilt unter anderem 

\begin{align}
\vec{AB} + (- \vec{AB}) &= \vec{AB} + \vec{BA} \\
                        &= \vec{AA}\\
                        &= \vec{0}
\end{align}

### Distributivgesetz und Skalierung von Vektoren

Vektoren können nicht nur Addiert und Subtrahiert werden, sondern auch mit Skalaren in ihrem Betrag verändert. Es gelten die folgenden Sätze

Gegeben seien die Vektoren $\vec{a}$ und $\vec{b}$ sowie $\alpha$ und $\beta$ $\in \mathbb{R}$. Es gilt
\begin{align}
(\alpha + \beta)\vec{a} &= \alpha \vec{a} + \beta\vec{a}\\
(\alpha  \beta)\vec{a} &= \alpha ( \beta\vec{a})\\
\alpha  (\vec{a}+\vec{b}) &=\alpha  \vec{a}+ \alpha  \vec{b}
\end{align}

### Einheitsvektor

Ein Einheitsvektor wird über seinen Betrag definiert, wenn

\begin{align}
|\vec{a}|&= 1
\end{align}

gilt, ist $\vec{a}$ ein Einheitsvektor. Einheitsvektoren werden nach $$\vec e_a = \frac{\vec a}{|\vec a|}$$ berechnet. 

### Unabhängigkeit von Vektoren 

Vektoren sind genau dann unabhängig voneinander wenn 

\begin{equation}
\vec{0} = \Sigma a_i \vec{x_i} 
\end{equation}

mit $i \in \mathbb{N}$ und $a_i = 0$ als _einzige_ Lösung gilt. 

Vereinfacht ausgedrückt, Vektoren sind unabhängig, wenn sie nicht als Linearkombination voneinander darstellbar sind, oder der Nullvektor durch eine gewichtete Summation erzeugbar ist. 

### Orthogonalität

Zwei vom Nullvektor verschiedene Vektoren sind genau dann orthogonal, wenn $$\vec a \cdot \vec b = 0$$ gilt. 

In Komponentendarstellung heist dies $$\vec a \cdot \vec b = \begin{pmatrix}a_x\\a_y\\a_z\end{pmatrix} \cdot \begin{pmatrix}b_x\\b_y\\b_z\end{pmatrix} = a_xb_x+a_yb_y+a_zb_z =0$$.

Der hier verwendete Operator $()\cdot()$ heißt Punktprodukt. 

## Koordinatendarstellung 

Bisher wurden Vektoren rein als gerichtete Größen im Raum betrachtet, aber nur bedingt mit Werten versehen die sie im Bezug zu einem Koordinatensystem setzen. 

Das gängigste Koordinatensystem ist hierbei das kartesische Koordinatensystem. Dieses wird von unabhängigen Einheitsvektoren aufgespannt. In einem zweidimensionalen System spannen sie eine Ebene in $\vec{x}$ und $\vec{y}$ auf, wobei die Achsen 90$^°$ zueinander gedreht sind. 
In einem dreidimensionalen System wird ein dritter unabhängiger Einheitsvektor $\vec{z}$ hinzugefügt, welcher im 90$^°$ Winkel auf der von $\vec{x}$ und $\vec{y}$ aufgespannten Ebene steht. 

Um Vektoren nun in Koordinaten darzustellen werden sie in ihere Komponenten zerlegt, die parrallel den Einheitsvektoren des Koordinatensystems liegen. So entspricht ein Vektor der zwei Einheiten in $x$ Richtung und drei in $y$ Richtung verläuft dem Vektor 

\begin{equation}
\begin{pmatrix}
2\\
3
\end{pmatrix}
\end{equation}

Wird diesem Vektor ein Betrag von in der $z$ Achse zugewiesen gilt folgende Notation

\begin{equation}
\begin{pmatrix}
2\\
3\\
1
\end{pmatrix}
\end{equation}

für den Fall, dass $\vec{a} \in \mathbb{R}^2$ (Ebene) gilt

\begin{equation}
\vec{a} = 
\begin{pmatrix}
a_x\\
a_y
\end{pmatrix}
\end{equation}

und abschließend gilt für $\vec{a} \in \mathbb{R}^3$ (Volumen)

\begin{equation}
\vec{a} = 
\begin{pmatrix}
a_x\\
a_y\\
a_z
\end{pmatrix}
\end{equation}

Verallgemeinert gilt damit für einen Vektor $\vec{a} \in \mathbb{R}^n \wedge n \in \mathbb{N}$

\begin{equation}
\vec{a} = 
\begin{pmatrix}
a_0\\
a_1\\
...\\
a_n
\end{pmatrix}
\end{equation}

Wobei der Index $a_i$ jeweils die zum Wert zugehörige Achse darstellt. 

## Vektor Betrag

Bisher wurde an mehreren Stellen der Betrag eines Vektors erwähnt, jedoch nicht erläutert wie dieser berechnet wird. Dargestellt in kartesischen Koordinaten kann der Betrag eines Vektors $\vec{a}$ in $\mathbb{R}^2$ nach 

\begin{equation}
|\vec{a}| = \sqrt{a_x^2+a_y^2}
\end{equation}

bestimmt werden. Der geneigte Leser mag hier den Satz des Pythagoras wiederzufinden. 

Analog berechnet sich der Betrag eines Vektors $\vec{a}$ in $\mathbb{R}^3$ nach

\begin{equation}
|\vec{a}| = \sqrt{a_x^2+a_y^2 + a_z^2}
\end{equation}

# Gerader Stoß

In den bisherigen Betrachtungen wurde angenommen, dass Körper, wenn sie sich berühren entweder in Ruhe sind (Lagerreaktionen, Statik) oder durch eine Kinematische Bindung ihre relative Lage zueinander nicht ändern können. Leicht ersichtlich ist dies jedoch nicht immer der Fall, wenn zwei Massen aufeinander treffen wird dies als Stoß bezeichnet. Der Stoß besitzt nicht nur akademische Relevanz, sondern auch solche für die Technik, als auch für die Physiksimulation in Videospielen. 

Stöße können grundsätzlich in zwei Kategorien unterteilt werden

* elastisch 
* plastisch

Die Unterscheidung dieser beiden Kategorien erfolgt nach der Änderung der Energie im System. Wird keine Energie verloren handelt es sich um einen elastischen Stoß, wird die gesamte Energie des Systems verloren handelt es sich um einen vollkommen plastischen Stoß. 

In der Realtät sind keine Prozesse verlustfrei, weswegen kein Stoß ideal elastisch sein kann, stattdessen sind Stöße stets auf einem Spektrum zwischen (nahezu ideal) elastischen Stoß, teil plastischen Stoß und vollplastischen Stoß angesiedelt. 

Im Folgenden soll der Stoß zwischen den Massen $m_1$ und $m_2$ mit den Geschwindigkeiten $v_1$ und $v_2$ betrachtet werden. Der Stoßvorgang selbst, soll aufgrund der hohen Komplexität, nicht betrachtet werden. Stattdessen wird der Zustand unmittelbar vor und nach dem Zusammentreffen der Massen betrachtet. 

Werte nach dem Stoß werden im Folgenden durch ein hochgestelltes $()^*$ von den Ausgagszuständen unterschieden. 

### Aufgabe

Im plastischen Stoß wird Energie vom betrachteten Massepunktsystem abgegeben. Welche Formen von Energie sind hier möglich?

## Gerader zentrischer Stoß

Zunächst soll der einfachste Fall des Stoßes behandelt werden. Während des elastischen Stoßes geht keine Energie verloren. Ein Stoß wird als gerade bezeichnet, wenn die Geschwindigkeitsvektoren $\vec v_1, \vec v_2$ der betrachteten Massen $m_1, m_2$ vor dem Stoß in einer Linie liegen, bzw. die Geschwindigkeitsvektoren normal auf der Berührungsebene stehen. 

Durch die Einschränkung, dass der Stoß gerade ist, kann für die folgende Betrachtung mit den Beträgen der Geschwindigkeitsvektoren $|\vec v_i| = v_i$ gearbeitet werden. 

Aus den bisherigen Betrachtungen der Energierhaltung eines Massepunktsystems wissen wir, dass das die Summe der Energien im System konstant ist. Gleichermaßen ist der Impuls des Gesamtsystems konstant. Diese Eigenschaften können wir uns zunutze machen um das Verhalten vor und nach dem Stoß zu modellieren. 

Vor dem Stoß der beiden Massen $m_1, m_2$ kann der Gesamtimpuls des Systems durch 

$$
p = \sum_i m_i v_i = m_1 v_1 + m_2 v_2 
$$

beschrieben werden. 

Der Gesamtimpuls nach dem Stoß kann in gleicher Form durch 

$$
p^* = \sum_i m_i v_i^* = m_1 v_1^* + m_2 v_2^*
$$

beschrieben werden. 

Durch den Impulserhaltungssatz gilt 

$$p = p^*$$

und somit nach umstellen 

$$ m_1 (v_1 - v_1^*) = m_2 (v_2^* - v_2) $$

als Zusammenhang der Geschwindigkeiten vor und nach dem Stoß. 

Mit der Energie als Ausgangsgröße für die Betrachtung des Stoßes kann geschrieben werden 

$$ E_k = \sum_i E_{k,i} = \sum_i m_i v_i^2 = m_1 v_1^2 + m_2 v_2^2 $$

als Beschreibung der kinetischen Energie vor dem Stoß, nach dem Stoß gilt 

$$ E_k^* = \sum_i E_{k,i}^* = \sum_i m_i v_i^{*2} =  m_1 v_1^{*2} + m_2 v_2^{*2} $$. 

Nimmt man den Energieerhaltungssatz zur Hilfe kann 

\begin{align}
E_k &= E_k^* \\ 
m_1 v_1^2 + m_2 v_2^2 &= m_1 v_1^{* 2} + m_2 v_2^{* 2} \\
 m_1 (v_1^2 - v_1^{* 2}) &= m_2 (v_2^{* 2} - v_2^2)
\end{align}

geschrieben werden. Die Energie vor und nach dem Stoß ist konstant. 

Setzt man beide Gleichungen ineinander ein, folgt schließlich 

$$v_1 - v_2 = v_2^* - v_1^*$$

als Ausdruck, welcher auch als Stoßbedingung beschrieben wird. Die Stoßbedingung besagt; 

<center>_Die Geschwindigkeitsdifferenz vor dem Stoß ist gleich der Differenz nach dem Stoß._</center>

Setzt man die Stoßbedingung in die zuvor gefundenen Zusammenhänge ein, ergeben sich für die Geschwindigkeiten nach dem Stoß folgende Zusammenhänge:

$$
v_1^* = \frac{m_1 v_1 + m_2 (2 v_2 - v_1) }{m_1 + m_2}
$$

und für $v_2^*$

$$
v_2^* = \frac{m_2 v_2 + m_1 (2 v_1 - v_2) }{m_1 + m_2}
$$

## Gerader (teil-)plastischer Stoß

Zuvor wurde der Fall behandelt, dass während des Stoßes keine Energie verloren geht. Diese Annahme wird im folgenden verworfen. Zur Modellierung des Energieverlustes wird die Newtonsche Stoßhypothese herangezogen welche als 

$$
v_2^* - v_1^* = e (v_1 - v_2)
$$

definiert ist. Hierbei ist die Stoßzahl $e$ definiert als $e \in \mathrm{N} \; | \; 0 \leq e \leq 1 $. Diese gibt an wie viel Energie während des Stoßes verloren geht. Bei $e=1$ handelt es sich um einen verlustfreien, elastischen, Stoß, bei $e=0$ um einen vollplastischen Stoß. 


Erweitert man die zuvor gefundenen Zusammenhänge für die Geschwindigkeiten $v_1^*, v_2^*$ folgen 

\begin{align}
v_1^* &= \frac{m_1 v_1 + m_2 v_2 + e m_2 (v_2 - v_1) }{m_1 + m_2} \\
v_2^* &= \frac{m_2 v_2 + m_1 v_1 + e m_1 (v_1 - v_2) }{m_1 + m_2}
\end{align}

Der Energieerhaltungssatz gilt grundsätzlich weiterhin, die Energie geht dem System zwar als Bewegungs- oder Lageenergie verloren, existiert als nicht verwendbare Verlustenergie $E_v$ jedoch weiterhin. 

In Gleichungen Ausgedrückt ist die Verlustenergie gleich der Differenz der Energien vor und nach dem Stoß mit  

$$
E_v = E_{kin} - E_{kin}^*
$$

mit den Ausdrücken für die jewaigen kinetischen Energiene gilt damit
$$
E_v = m_1 v_1^2 + m_2 v_2^2 - m_1 v_1^{* 2} - m_2 v_2^{* 2}
$$

und mit den zuvor gefundenen Zusammenhängen für die Geschwindigkeiten nach dem Stoß folgt schließlich

$$
E_v = \frac{1-e^2}{2} \frac{m_1 m_2}{m_1 + m_2}(v_1-v_2)^2
$$. 

Der Grenzfall für den vollkommen plastischen Stoß mit $e=0$ ist mit 
$$
E_v = E_{v, max} \frac{1}{2} \frac{m_1 m_2}{m_1 + m_2}(v_1-v_2)^2
$$

und ist gleich der maximalen Energie die in einem Stoß verloren gehen kann. 

Die resultierende Geschwindigkeit der beiden Massen ist hiernach über 

$$
v_1^* = v_2^* = \frac{m_1 v_1 + m_2 v_2}{m_1 + m_2} = v^* 
$$

gegeben.

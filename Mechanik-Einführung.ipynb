# Einführung in die Mechanik

Um die Dynamischen Gleichungen vorzubereiten sollen zunächst statische Probleme betrachtet werden um ein Grundverständnis von Kräften zu erlangen, sowie den Umgang mit Kräftegleichgewichten und der Lösung von Gleichungen zu zu gewinnen. 

## Modellbildung

Bevor wir in die formale Beschreibung der Newton'schen Mechanik einsteigen sollte man sich eine Frage stellen:
    - Was zeichnet ein gutes Modell aus?
Diese Frage ist aus zwei Blickwinkeln relevant; Zum einem stellt sich die Frage: 'Wie akkurat stellt mein System die Realität dar?' auf der anderen Seite 'Wie groß ist der Rechenaufwand wenn ich noch genauer in meiner Beschreibung werde?', denn am Ende gilt immer Anderson's Law: 

<center>_"There is no problem no matter how complicated, that when viewed from the right angle, doesn't become even more complicated"_</center>

Mit diesen Fragen, und dem Umstand das man ein Modell beliebig komplex bauen kann, ist es möglich zu einer einfachen Frage zu kommen die man seinem Modell stellen kann
- Ist mein Modell genau genug?
Denn hier ist kommt man in aller Regel zum einfachsten möglichen Modell, welches die Realität abbildet. 

##### The Catch

Dieser Kurs beschäftigt sich nicht nur mit der klassischen Mechanik, sondern mit der klassischen Mechanik im Kontext von Videospielen. Hierdurch müssen auch einige andere Parameter mit betrachtet werden und auch anders gewichtet. Auf eine Frage reduziert kann man sein Modell fragen 
- Ist mein Modell genau genug, ist es performant und vor allem _macht es dem Spieler Spaß?_

Die Aspekte sind schnell begründet, dass physikalische Modell soll die Realität insofern abbilden damit sich der Spieler intuitiv mit der Welt interagieren kann. Es muss darüber hinreichend Performant sein um nicht zu viele CPU oder GPU Zyklen zu verwenden und so die Framerate zu riskieren. Aber vor allem: kein System sollte dem Spielererlebnis im Weg stehen. Hier lose mit den Regeln der Natur zu spielen um das Spiel interessanter zu gestalten ist gute Praxis. 

##### The other Catch

Der Spieler bemerkt vermutlich wesentlich weniger von dem, was in der Simulation akkurat dargestellt wird. Manchmal hilft es bereits die Präsentation zu verändern ohne die Simulation anzufassen um ein bessere Spielererleben zu erhalten. [Dieser GDC-Beitrag von Jared Cone](https://www.youtube.com/watch?v=ueEmiDM94IE) ist hierzu recht sehenswert. 

## Wann gilt die klassiche Mechanik? 

Mit dem Aufkommen der Relativitätstheorie sollte man meinen, dass die klassiche Mechanik keine Relevanz mehr hat, immerhin ist sie rund 300 Jahre alt. Aber hier sollte man nicht vorschnell handeln. 

Die Relativitätstheorie ist vor allem relevant wenn: 

* man sich sehr schnell bewegt (deutlicher Teil der Lichtgeschwindigkeit)
* man sehr groß ist
* sehr große Strecken zurücklegt
* in der Nähe eines schwarzen Lochs ist

... sprich, in keinem Fall der einem im Alltag begegnet.

Aus diesem Grund wird sich hier mit der klassischen Mechanik befasst, da sie in der Lage ist die Natur in sehr guter Art abzubilden. 

## Kräfte

Kräfte wirken quasi immer, auf alles, und meistens sind wir uns ihnen noch nicht einmal bewusst. Sie besorgen die Bewegung von Objekten, oder verhindern es und können in drei Gruppen unterteilt werden:

* Äußere Kräfte
* Innere Kräfte
* Reaktionskräfte

Äußere Kräfte sind Kräfte die in ein System eingeprägt werden. Innere Kräfte sind Reaktionen die innerhalb einer Struktur entstehen und werden zu einem soäteren Zeitpunkt näher betrachtet. Die Dritte Gruppe sind die Reaktionskräfte, sie entstehen zwischen Teilstrukturen oder in Lagern. 

Die einfachste äußere Kraft ist hierbei vermutlich die Gewichtskraft $G$. Sie wird durch

\begin{equation}
G = m g
\end{equation}

definiert, wobei $m$ die Masse eines Objekts ist und $g$ die Erdbeschleunigung. 

Die Grundeinheit von Kräften ist Newton (N), benannt nach Sir Isaac Newton, dem Erstbeschreiber der Klassischen Mechanik. 

## Die mechanische Kartoffel. 

Wir wissen nun das Kräfte mit Bewegung zusammenhängen, aber wie genau? 

Zunächst betrachten wir das Objekt, auf welches die Kräfte wirken. Kräfte können zunächst auf jedes Objekt wirken, welches eine Masse $m$ besitzt. Die Masse kann nach

\begin{equation}
m = \rho V
\end{equation} 

mit dem Volumen $V$ und der Massedichte $\rho$ bestimmt werden. 

Diese Masse kann dann in einem Masseschwerpunkt durch einen Ortsvektor eindeutig beschrieben werden. 

Das restliche Volumen ist weiterhin vorhanden. Hier nehmen wir die Rahmenbedingungen eines starren Körpers an, bzw. der mechanischen Kartoffel. 

Die mechanische Kartoffel ist eine Idealisierung. Sie kann sich durch äußere Kräfte nicht verformen, wodurch Kräfte auf ihrer Wirkungslinie frei verschoben werden können. 

Hieraus ergibt sich sofort das erste Axiom, das Linienflüchtigkeitsaxiom:

"Kräfte können ohne Änderung ihre Wirkung auf einen starren Körper auf ihrer Kraftwirkungslinie verschoben werden."

Es ist einfach ersichtlich, dass bei einem verformbaren Körper der Angriffspunkt einer Kraft von großer Bedeutung ist, da die Änderung der Form durch die äußeren Kräfte zu einer Änderung der geometrischen Zusammenhänge führt. Idealisieren wir diese Verformung jedoch weg, entfällt das Problem. 

Das nächste Axiom ist das Gleichgewichtsaxiom 

"Zwei Kräfte sind im Gleichgewicht wenn sie betragsgleich sind und antipatallel verlaufen."

Dies ergibt sich schnell wenn man ein Kräfteparrallelogramm zur Hilfe nimmt. Trägt man hier zwei Vektoren mit den Oben genannten Eigenschafte ein, ergbit sich der Nullvektor, oder anders ausgedrückt gilt:

\begin{equation}
\vec{F_1} + \vec{F_2} = \vec{0}
\end{equation}

## Freischnitt und Wechselwirkungsgesetz

Der Satz 

<center> _"actio est reactio"_ </center>

ist recht bekannt, aber was bedeutet er genau? 

Um das Wechselwirkungsgesetz zu verstehen ist es nützlich, sich mit dem Konzept des Freischnitts, auseinanderzusetzen. Bei einem Freischnitt wird die mechanische Kartoffel von ihrer Unterlage losgelöst und man betrachtet die Kräfte die von Lager und dem aufliegenden Körper ausgehen. 

So wirkt, wenn man auf dem Boden steht, eine Kraft die vom Körper Richtung Boden zeigt, und eine Reaktionskraft, die einen davon abhält durch den Boden zu fallen. Idealerweise ist diese Eigenschaft in Spielen ebenfalls implementiert. 

Aber auch ohne direkten Kontakt gilt dieses Gleichgewicht. Betrachtet man das System von Erde und Mond ist die Gravitationskraft die vom Mond ausgeht genauso groß und entgegengesetzt der Kraft mit welcher die Erde den Mond anzieht. 



## Translatorische Kräfte

Im einfachsten Fall greift eine Kraft direkt im Massenschwerpunkt an, oder kann durch Anwendung des Linienflüchtigkeitsaxioms in den Schwerpunkt verschoben werden. Eine im Schwerpunkt wirkende Kraft führt zu einer Verschiebung (Translation) entlang ihrer Wirkachse ohne den Lagewinkel zu verändern. 

## Kräftezerlegung

Kräfte treten nicht in allen Fällen parallel zu den Achsen des Koordinatensystems auf. Um sie denoch in Gleichungen behandeln zu können, ist es möglich, sie in ihre Komponenten zu zerlegen. Die einzelnen Komponenten verlaufen entlang der Achsen des gewählten Koordinatensystems. Im ebenen Fall die $x$ und $y$-Achsen. Für die Projektion einer Kraft auf diese Achsen werden die trigonometrischen Funktionen $sin$ und $cos$ verwendet. 

Die am Winkel anliegende Komponente ergibt sich nach 

\begin{equation}
F_a = F cos(\phi) 
\end{equation}

die gegenüberliegende Komponente entsprechend nach

\begin{equation}
F_g = F sin(\phi)
\end{equation}

wobei $F_g$ hier nicht mit der Gewichtskraft $G$ verwechselt werden soll. 

Je nach Lage der Achsen bzw. des gegebenen Winkels können die Kräfte $F_a,F_g$ den Achsen $x$ oder $y$ entsprechen. 



## Momente

Bisher wurde angenommen, dass jede angreifende Kraft in ihrer Wirkungslinie durch den Masseschwerpunkt geht. Oder in anderen Worten, dass eine Kraft nur zu einer Verschiebung führt, jedoch nicht zu einer Drehung.

Wenn eine Kraft mit einem Abstand $l$ zum Masseschwerpunkt an einem starren Körper angreift bildet sie ein Moment $M$ nach

\begin{equation}
M = Fl
\end{equation}

aus. Hierbei muss die Kraft $F$ zwingend im rechten Winkel der Länge $l$ stehen, damit diese Gleichung gilt. 

Dies ist jedoch in der Praxis nicht immer der Fall. Um das resultierende Moment einer beliebigen Kraft auf einen beliebigen Punkt zu berechnen,muss diese in seine Komponenten zerlegt werden. Hierfür sucht man die Projektion einer Kraft $F$, mit dem Angriffspunkt $(x,y)$ und einem Winkel $\phi$ den starren Körper angreift, auf die $x$ und $y$-Achse. Es wird das Moment $M_0$ im Bezug zum Ursprung gesucht. 

Hier ergibt sich,

\begin{align}
F_x = F cos(\phi)
F_y = F sin(\phi)
\end{align}

Das Gesamtmoment kann auch als Summe aller Teilmomente bestimmt werden nach 

\begin{equation}
M_{gesamt} = \Sigma M_i
\end{equation}

womit folgt 

\begin{equation}
M_{0} = - F_x  y + F_y  x = F (x sin(\phi) - y cos(\phi))
\end{equation}

Alternativ kann man anstatt der Kräftezerlegung auch versuchen einen analytischen Begriff für den Hebelarm $l$ zu finden. 

Dieser Hebelarm kann mit eine wenig Geometrie konstruiert werden 



Und ergibt sich damit nach

\begin{equation}
l = x sin(\phi) - y cos(\phi)
\end{equation}

setzt man diesen Hebelarm ein folgt wiederum 

\begin{equation}
M_{0} = F (x sin(\phi) - y cos(\phi))
\end{equation}

womit wir das gleiche Moment berechnet haben. Anzumerken ist, dass bei der ersten Methode auf die Richtung der projizierten Kraft zu achten ist, während die Vorzeichen der Kraftkomponenten beim ermitteln des Hebelarms als Nebenprodukt anfallen und nicht separat ermittelt werden müssen. 

## Haftreibung

Jeder der schonmal versucht hat eine schwere Kiste zu verschieben hat schnell feststellen müssen, dass diese auch ohne am Boden veranktert zu sein eine Gegenkraft ausübt. Auf der anderen Seite ließe sich diese Kiste vermutlich auf Glatteis leichter schieben, sofern man nicht selbst ausrutscht. 

Was geschieht hier genau? 

Jedes Objekt das mit einer Oberfläche ein anderes Objekt berührt erzeugt im Berührungspunkt eine Normalkraft $N$ die senkrecht auf der Berührungsfläche steht. Des Weiteren haben die Flächen immer einen Reibungswert $\mu$ welcher durch die Oberflächenrauhigkeit bestimmt wird.

Die Haftreibung $R_H$ ist das Produkt aus Normalkraft $N$ und dem Reibungswert $\mu$ nach

\begin{equation}
R_H = N \mu_H
\end{equation}



Bei der Haftreibung handelt es sich ebenfalls um eine Reaktionskraft. Wird sie überschritten kommt der Körper ins gleiten. Hierbei ist wichtig, dass die Gleitreibung in aller Regel wesentlich geringer ist als die Haftreibung. In anderen Worten: wenn es rutscht, dann richtig. 

Gut zu beobachten bei Autofahrern die es mit dem Gaspedal und Reifenverschleiß nicht so eng sehen.

# Massepunkt Systeme - Fortsetzung

In [den Grundlagen der Dynamik](Mechanik-Dynamik.ipynb) wurden die Grundlagen des Massepunktes bereits beschrieben. Hier sollen nun weitere Eigenschaften aufgezeigt werden, die notwendig sind um den Stoß zu behandeln. 

Wie bereits besprochen können zwei verschiedene Arten der Bindung zwischen zwei Massepunkten auftreten:

* Kinematische Bindungen
* Physikalische Bindungen

In kinematischen Bindungen ist die Lage der Massepunkte durch eine feste Geometrie beschrieben, die eindeutige Zusammenhänge für eine Bewegungsgleichung liefert. 

Bindungen grenzen grundsätzlich den Freiheitsgrad eines Punktes ein, eine unabhängige Bewegung einzelner Massepunkte ist nicht möglich. In anderen Worten: 

<center>
Die Massen eines gekoppelten Systems führen stets voneinander abhängige Bewegungen aus. 
</center>

Der Freiheitsgrad eines Systems wird mit $f$ gekennzeichnet. 

Grundsätzlich ergibt sich $f$ nach 

\begin{equation}
f = 3n - d
\end{equation}

Wobei $n$ die Anzahl der Massepunkte in einem System darstellt und $d$ die Anzahl der Bindungen. Die $3n$ ergeben sich aus der Betrachtung des dreidimensionalen Systems. 

Der ebene Fall ergibt sich zu 

\begin{equation}
f = 2n - d
\end{equation}

und schließlich der einachsige Fall 

\begin{equation}
f = n - d
\end{equation}

In jedem dieser Fälle wird nur die Translation betrachtet. 

### Beispiel einachsiges System

Was bedeutet einachsige Bewegung mit gekoppelten Massen genau? Das folgende Beispiel soll dies im Detail illustrieren. 

Nehmen wir an, dass zwei Massen mit einem dehnssteifen, aber biegeschlaffen, Seil miteinander verbunden sind (bedeutet, dieses Seil kann beliebigen Zug, jedoch keinen Schub aufnehmen). 

hiermit ergibt sich 

\begin{align}
n &= 2 \\
d &= 1\\
f &= 2-1 
\end{align}

der Freihetisgrad $f$ = 1. Mit den gewählten lokalen Koordinaten $x_1$ und $x_2$ kann die kinematische Beziehung $$x_1 = x_2$$ aufgestellt werden. 

Setzt man diese Beziehung in die Bewegungsgleichungen der Massen $m_1$ und $m_2$ ein, kann das System vollständig mit $x_1$ oder $x_2$ beschreieben werden. 

### Ebene Bewegung 

Entgegen der einachsigen Bewegung stehen in der ebenen Bewegung zwei Freiheitsgrade zur verfügung. Eine kinematische Bindung wird hier in der Regel durch eine masselose Stange dargestellt. Diese kann sowohl Zug als auch Schub aufnehmen. 

### Beispiel Ebene Bewegung 

Im Folgenden sind zwei, durch eine masselose Stange der Länge $l$, verbundene Massepunkte dargestellt. Der Freiheitgrad ergibt sich nach 

\begin{align}
n &= 2 \\
d &= 1\\
f &= 2 * 2-1 = 3
\end{align}

Die Lagebeziehung lässt sich aus der allgemeinen Kreisgleichung 

$$x^2 + y^2 = l^2$$ 

in der Normalform herleiten. Hierbei sind $x$ und $y$ die Koordinaten eines beliebigen Punktes auf einem Kreis mit dem Radius $l$. 

Angewendet auf diesen Fall, folgt die kinematische Bindung mit 

$$(x_2 - x_1)^2 + (y_2 - y_1)^2 = l^2$$ 

um die Abhängigkeit der Lage der Massepunkte zu beschreiben. 

Die Lage diese Körpers muss über drei Variablen abgebildet werden um diese vollständig zu beschreiben. Dies ist durch zwei Koordinaten $x,y$ sowie einen Drehwinkel $\phi$ möglich. Alternativ kann die Kreisgleichung nach einer der vier Koordinaten aufgelöst werden und zur Substitution verwendet werden. 

## Äußere und Innere Kräfte von Massepunktsystemen

In der Einführung bereits angesprochen können Kräfte in drei Gruppen unterteilt werden. 

* Äußere Kräfte
* Innere Kräfte
* Reaktionskräfte

Äußere Kräfte und Reaktionskräfte wurden bereits in vorherigen Kapiteln behandelt. Hier sollen nun die inneren Kräfte betrachtet werden. 

Innere Kräfte sind solche Kräfte, die nach außen hin nicht sichtbar sind und durch einen Freischnitt sichtbar gemacht werden müssen. Eine innere Kraft ist beispielsweise jene Kraft, welche durch einen Stab der zwei Massepunkte verbindet, geleitet wird. Hierbei wird an den Schnittufern eines Freischnitts eine Normalkraft $N$ eingetragen. Wird dieses Kräftepaar aufsummiert verbleibt keine resultierende Kraft. 

Innere Kräfte sind den Kräften in einem statischen System vergleichbar, sie können wie die Lagerreaktionen nur gezeigt werden wenn ein Freischnitt vorgenommen wird und es gilt die Gleichgewichtsbedingung $\sum F = 0$. 

Formal gilt für Normalkräfte in einem Freischnitt damit $$N_{ij} = - N_{ji}$$. Die Indices $i,j$ beschreiben hierbei die am Freischnitt beteiligten Massepunkt. Der erste Index zeigt die Seite des Schnittufers, der zweite Index die gegenüberliegende Seite.

Stellt man die newtonschen Grundgleichung für das Gesamtsystem auf folgt 
\begin{equation}
    m_i \vec a_i = m_i \vec{ \ddot r_i} = \vec F_i + \sum \vec N_{ij}
\end{equation}
wobei $m_i$ die einzelnen betrachteten Massen, $\vec a_i$ und $\vec r_i$ die zugehörigen Beschleunigungen und Ortsvektoren sind. Die $\vec F_i$ beschreibt die wirkenden äußeren Kräfte. 

## Betrachtung des Gesamtsystems

Im Folgenden soll eine Beschreibung für das Gesamtsystem gefunden werden, welches sich aus den einzelnen kinematisch verbundenen Massen ergibt. 

Hierfür soll zunächst über die Einzelmassen aufsummiert werden mit 

$$\sum_i m_i \vec a_i = \sum_i \vec F_i + \sum_j \sum_i \vec N_{ij}$$. 

Unter Berücksichtigung der Randbedingung dass $N_{ij}=-N_{ji}$ ist, wird die Doppelsumme $\sum_j \sum_i \vec N_{ij}$ zu Null und es ergibt sich 

$$\sum_i m_i \vec{\ddot{r_i}} = \sum_i \vec F_i$$

für das Gesamtsystem.

Dieser Zusammenhang verdeutlicht, dass die inneren Kräfte keine Auswirkungen auf das Verhalten des Gesamtsystems besitzen. 

Hier ist jedoch der Ausdruck $\sum_i m_i \vec{\ddot{r_i}}$ recht unhandlich, um diesen Ausdruck zu vereinfachen, soll mittels des Schwerpunktsatzes, das System der Massepunkte auf einen Massepunkt im Schwerezentrum vereinfacht werden, sowie der Ortsvektor des Schwerpunktes ermittelt werden. 

Der Ortsvektor des Schwerpunkts ergibt sich nach $$ \vec{ r_s} = \frac{\sum_i m_i \vec{ r_i}}{\sum_i m_i} $$ als über die Einzelmassen $m_i$ gewichtetes Mittel der Ortsvektoren $\vec r_i$. 

Mit $$m = \sum_i m_i$$ vereinfacht sich der Ausdruck zu $$ \vec{ r_s} = \frac{1}{m}\sum_i m_i \vec{ r_i} $$ womit wir schließlich zum Ausdruck $$ \sum_i m_i \vec{\ddot r_s} = m  \vec{\ddot r_s} = m \vec a_s$$

Es gilt weiterhin die newtonsche Grundgleichung $$m \vec a_s = \vec F$$, wobei $$\sum_i \vec F_i = \vec F$$ als resultirende äußere Kraft definiert ist. 

Diese Zusammenhänge könne wie gewohnt in die kartesischen Koordinaten zerlegt werden nach 

\begin{align}
m \ddot x_s &= F_x \\
m \ddot y_s &= F_y \\
m \ddot z_s &= F_z
\end{align}

wobei $m$ weiterhin die Gesammasse darstellt, $\ddot x_s, \ddot y_s, \ddot z_s$ die Schwerpunktbeschleunigungen und $F_i$ die jewaigen Kraftkomponenten. 

### Gesamtimpuls und Impulserhaltungssatz

Der Impuls $\vec p$ eines Massepunktsystems folgt als Überlagerung der Einzelimpulse $\vec p_i$. Hiermit folgt der Zusammenhang 

$$\vec p = \sum_i \vec p_i = \sum_i m_i \vec v_i$$. 

Unter Zuhilfenahme der zuvor gefundenen Zusammenhänge gilt damit 

$$\vec p = m \vec{\dot r_s} = m \vec v_s$$ 

womit der Impuls des Systems als Produkt von Gesamtmasse und Geschwindigkeit des Schwerpunktes $v_s$ beschrieben werden kann. 

### Gesamtdrall 

Der Drall $\vec L$ des Gesamtsystems ergibt sich weiterhin entsprechend des Drallsatzes 

$$\vec L_i = \vec r_i \times \vec p_i= \vec r_i \times m_i \vec{\dot v_i} $$, 

welcher den Drall $\vec L_i$ der Einzelmassen beschreibt, als Summe über alle Massen nach 

$$\vec L = \sum_i \vec L_i$$.

Ohne einwirkende äußere Momente ist der Drall konstant, es gelten die Zusammenhänge 
$$
\vec{\dot L} = \vec M
$$

wenn 

$$\vec M = 0 $$

folgt 

$$
\vec L = konst. 
$$

### Energieerhaltung

Erwatrungsgemäß gilt für ein Massepunktsystem genauso wie für einzelne Massepunkte der Energieerhaltungssatz nach 

$$E_{pot} + E_{kin} = konst.$$ 

wobei $E_{pot}$ die potenzielle Energie des Systems beschreibt und $E_{kin}$ die kinetische Energie. Hierbei wird angenommen, dass keine Reibung im System vorliegt. 

Die kinetische Energie des Systems ergibt sich nach 

$$E_{kin} = \sum \frac{m_i v_i^2}{2} = \sum \frac{m_i \dot r_i^2}{2}$$ 

als Superposition der einzelnen kinetischen Energien. Gleichermaßen gilt für die Lageenergie 

$$E_{pot} = \sum \frac{m_i g h_i}{2}$$ 

wobei $g$ die Gravitationsbeschleunigung und $h_i$ die Höhe im Schwerefeld zu einer gewählten Nulllage definiert.
